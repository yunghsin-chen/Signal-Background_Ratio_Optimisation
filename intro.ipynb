## Poisson distribution Introduction

```python
# Poisson distribution
x_poisson = np.linspace(0, 50, 100)

def Poisson(m, x):
    poisson_list = []
    for i in x:
        poisson_list.append( (m**i)*np.exp((-1)*m)/gamma(i+1) )
    return poisson_list

# plot Poisson distribution
plt.plot(x_poisson, Poisson(16, x_poisson), label = 'Background')
plt.plot(x_poisson, Poisson(16+3*4, x_poisson), label = 'Signal')
plt.vlines(16+3*4, -0.0025, 0.105)
plt.xlabel('Background', fontsize = 15)
plt.ylabel('Signal', fontsize = 15)
plt.legend(loc = 'best', fontsize = 13)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.ylim(-0.005, 0.11)
```

## Error function (illustration)

```python
x_error = np.linspace(0, 10, 100)

def Error(x):
    error_list = []
    for i in x:
        error_list.append((erf(i-5)+1)/2)
    return error_list

# plot
plt.plot(x_error, Error(x_error), label = 'Error Function')
plt.xlabel(r'$Number\;of\;signal\;\sigma\;from\;noise\;background$', fontsize = 15)
plt.ylabel('Score', fontsize = 15)
plt.legend(loc = 'best', fontsize = 13)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
```

## Mapping function (illustration)

```python
x_map = np.linspace(0, 1, 100)

def Map(x, alpha):
    map_list = []
    for i in x:
        map_list.append(i**(alpha))
    return map_list

# plot
plt.plot(x_map, Map(x_map, 0.1), label = r'$y = x^{0.01}$')
plt.plot(x_map, Map(x_map, 0.3), label = r'$y = x^{0.1}$')
plt.plot(x_map, Map(x_map, 1), label = r'$y = x^{1}$')
plt.plot(x_map, Map(x_map, 3), label = r'$y = x^{10}$')
plt.plot(x_map, Map(x_map, 10), label = r'$y = x^{100}$')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)
plt.legend(loc = 'best', fontsize = 13)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
```
